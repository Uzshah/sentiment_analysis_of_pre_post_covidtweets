```python
import pandas as pd
import numpy as np
from datetime import datetime
import string
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('tweets_users_merged.csv', parse_dates = ['tw_created_at'])
```

```python
df.shape[0]
```

```python
df = df[~df['text'].str.contains('RT @|RT@')] 
df = df[df.verified==False]
```

```python
df.shape[0]
```

```python
import string
import re
def remove_urls(text):
  text = str(text)
  url_pattern = re.compile(r'https?://\S+|www\.\S+')
  text = url_pattern.sub(r'link', text)
  # Words with punctuations and special characters
  text = re.sub(r'[^\w\s]','',text)
  return text
```

```python
df['text'] = df['text'].apply(remove_urls)
```

```python
from datetime import datetime
def ch(x):
    if x.year<2020:
        return "T1"
    return "T2"

df['period2'] = df['tw_created_at'].apply(lambda x: ch(x))
```

```python
def chd2(x):
    if x.year<2020:
        return [1,0]
    return [0,1]
#df[['T1','T2','T3']] = df['ts'].apply(lambda x: pd.Series(chd(x)))
y = df['tw_created_at'].apply(lambda x: chd2(x))
```

```python
t1,t2 = [],[]
for t in y:
    t1.append(t[0])
    t2.append(t[1])
df['T11'] = t1
df['T22'] = t2
```

```python
dfgp2 = df
```

```python
eng_df = dfgp2[dfgp2.lang == 'en']
fr_df = dfgp2[dfgp2.lang == 'fr']
nl_df = dfgp2[dfgp2.lang == 'nl']
sv_df = dfgp2[dfgp2.lang == 'sv']
it_df = dfgp2[dfgp2.lang == 'it']
es_df = dfgp2[dfgp2.lang == 'es']
pt_df = dfgp2[dfgp2.lang == 'pt']
de_df = dfgp2[dfgp2.lang == 'de']
ca_df = dfgp2[dfgp2.lang == 'ca']

fr_df = fr_df.append(ca_df)
```

```python
gp2 = eng_df[['author_id','T11','T22']].groupby(['author_id']).sum()
gpm2 = gp2[(gp2.T11>0)&(gp2.T22>0)]
print(gpm2)
print()
print()

dfgp = pd.merge(left=df, right=gpm2, left_on='author_id', right_on='author_id')

print('this is confirmation that each user have tweet in both phase {}'.format(len(before.author_id.unique()) == len(after.author_id.unique())))

from MySenti import PySentiStr
senti = PySentiStr()
senti.setSentiStrengthPath('C:SentiStrength.jar') # Note: Provide absolute path instead of relative path
senti.setSentiStrengthLanguageFolderPath('C:/SentiStrength_for_python-master/')

dfgp['text'] = dfgp['text'].map(lambda x: str(x))
dfgp.dropna(subset=['text'], inplace=True)
list_tweet = dfgp['text'].to_list()
result = senti.getSentiment(list_tweet, lang='EN')
dfgp['sentiment'] = result

before = dfgp[dfgp.period2 == 'T1']
after = dfgp[dfgp.period2 == 'T2']

bfdf = before[['author_id', 'sentiment']].groupby('author_id').agg(lambda x:x.value_counts().index[0])
afdf = after[['author_id', 'sentiment']].groupby('author_id').agg(lambda x:x.value_counts().index[0])

indexes = afdf.index
df_new =pd.DataFrame()
df_new['author_id'] = indexes
df_new['T1'] = bfdf['sentiment'].to_list()
df_new['T2'] = afdf['sentiment'].to_list()
print(df_new.head())
print()
print()
print('Negative Negative {}'.format(len(df_new[(df_new['T1'] == -1) & (df_new['T2'] == -1)])))
# print('Negative Neutral {}'.format(len(df_new[(df_new['T1'] == -1) & (df_new['T2'] == 0)])))
print('Negative Positive {}'.format(len(df_new[(df_new['T1'] == -1) & (df_new['T2'] == 1)])))
# print('Neutral Negative {}'.format(len(df_new[(df_new['T1'] == 0) & (df_new['T2'] == -1)])))
# print('Neutral Neutral {}'.format(len(df_new[(df_new['T1'] == 0) & (df_new['T2'] == 0)])))
# print('Neutral Positive {}'.format(len(df_new[(df_new['T1'] == 0) & (df_new['T2'] == 1)])))
print('Positive Negative {}'.format(len(df_new[(df_new['T1'] == 1) & (df_new['T2'] == -1)])))
# print('Positive Neutral {}'.format(len(df_new[(df_new['T1'] == 1) & (df_new['T2'] == 0)])))
print('Positive Positive {}'.format(len(df_new[(df_new['T1'] == 1) & (df_new['T2'] == 1)])))
```
